Problem Setup (Transactions + Threshold Interpretation)
Requirement: “A database has five transactions”
We will encode the transactions as lists/sets.
From the image:

T100 = {M, O, N, K, E, Y}
T200 = {D, O, N, K, E, Y}
T300 = {M, A, K, E}
T400 = {M, U, C, K, Y}
T500 = {C, O, O, K, I, E} → treat as {C, O, K, I, E} (duplicates don’t change presence)

Requirement: “Let min sup = 60% and min conf = 80%”


With 5 transactions, support threshold count is:
0.60*5=30.60 \times 5 = 30.60*5=3
So any frequent itemset must appear in at least 3 transactions.

```python
#Install + Imports
!pip -q install mlxtend

import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpgrowth, association_rules
```

```python
#Encode Transactions

transactions = [
    ['M','O','N','K','E','Y'],      # T100
    ['D','O','N','K','E','Y'],      # T200
    ['M','A','K','E'],              # T300
    ['M','U','C','K','Y'],          # T400
    ['C','O','K','I','E']           # T500 (duplicate O removed)
]

te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
#Apriori Frequent Itemsets (min_support=0.60)

min_sup = 0.60

freq_ap = apriori(df, min_support=min_sup, use_colnames=True)
freq_ap['support_count'] = (freq_ap['support'] * len(transactions)).round().astype(int)
freq_ap.sort_values(['support','itemsets'], ascending=[False, True]).reset_index(drop=True)
```

```python
#Apriori Association Rules (min_confidence=0.80)

min_conf = 0.80

rules_ap = association_rules(freq_ap, metric="confidence", min_threshold=min_conf)

# add support counts for readability
rules_ap['support_count'] = (rules_ap['support'] * len(transactions)).round().astype(int)
rules_ap['antecedent_support_count'] = (rules_ap['antecedent support'] * len(transactions)).round().astype(int)
rules_ap['consequent_support_count'] = (rules_ap['consequent support'] * len(transactions)).round().astype(int)

rules_ap.sort_values(['confidence','lift','support'], ascending=[False, False, False]).reset_index(drop=True)
```

Interpretations:


M → K has confidence 1.0 but lift 1.0

Because K is always present (support(K)=1.0)
Meaning: even without M, K is guaranteed; so M doesn’t “increase” K.
Interpretation: “Whenever M occurs, K occurs”—true, but not very informative because K is universal.



O → E has confidence 1.0 and lift 1.25

support(E)=0.80, confidence=1.0
lift = 1 / 0.80 = 1.25
Interpretation: When O appears, E becomes 25% more likely than baseline.



K → E has confidence 0.80 and lift 1.0

baseline support(E)=0.80
so knowing K doesn’t change probability of E (K always occurs).
Interpretation: Rule passes confidence but is weak by lift.



O → KE has confidence 1.0 and lift 1.25

support(KE)=0.80; confidence=1.0
Interpretation: Every time O appears, it always comes with both K and E in this dataset.

```python
#FP-Growth Frequent Itemsets
freq_fp = fpgrowth(df, min_support=min_sup, use_colnames=True)
freq_fp['support_count'] = (freq_fp['support'] * len(transactions)).round().astype(int)
freq_fp.sort_values(['support','itemsets'], ascending=[False, True]).reset_index(drop=True)
```

```python
#FP-Growth Association Rules

rules_fp = association_rules(freq_fp, metric="confidence", min_threshold=min_conf)

rules_fp['support_count'] = (rules_fp['support'] * len(transactions)).round().astype(int)
rules_fp.sort_values(['confidence','lift','support'], ascending=[False, False, False]).reset_index(drop=True)
```

```python
#ECLAT (Vertical Format)

from itertools import combinations

tids = list(range(1, len(transactions)+1))  # 1..5
trans_sets = [set(t) for t in transactions]

# Build vertical representation: item -> set of TIDs
vertical = {}
for tid, items in zip(tids, trans_sets):
    for it in items:
        vertical.setdefault(it, set()).add(tid)

min_count = int(min_sup * len(transactions))  # 3

def eclat(vertical_db, min_count):
    # frequent itemsets stored as {frozenset(items): tidset}
    frequent = {}

    # Start with 1-item candidates
    items = sorted(vertical_db.items(), key=lambda x: x[0])
    level = [(frozenset([it]), tidset) for it, tidset in items if len(tidset) >= min_count]

    k = 1
    while level:
        next_level = []
        # record all itemsets in this level
        for itemset, tidset in level:
            frequent[itemset] = tidset

        # generate candidates by joining
        for i in range(len(level)):
            for j in range(i+1, len(level)):
                a, tid_a = level[i]
                b, tid_b = level[j]
                union = a | b
                if len(union) == k + 1:
                    inter = tid_a & tid_b
                    if len(inter) >= min_count:
                        next_level.append((union, inter))

        # remove duplicates
        unique = {}
        for itemset, tidset in next_level:
            unique[itemset] = tidset
        level = [(k, v) for k, v in unique.items()]
        k += 1

    return frequent

freq_eclat = eclat(vertical, min_count)

# Convert to DataFrame
rows = []
for itemset, tidset in freq_eclat.items():
    rows.append({
        "itemsets": set(itemset),
        "support_count": len(tidset),
        "support": len(tidset)/len(transactions),
        "tids": sorted(list(tidset))
    })
df_eclat = pd.DataFrame(rows).sort_values(['support','support_count'], ascending=False).reset_index(drop=True)
df_eclat
```
