Step 2 – Build Time‑Ordered Purchase Sequences per Customer

```python
customer_sequences = (
    df.sort_values("Purchase_Timestamp")
      .groupby("Customer_ID")["Item_Name"]
      .apply(list)
)

customer_sequences
```

Step 3 – PrefixSpan‑Style Sequential Pattern Mining

```python
from collections import Counter
```

```python
def extract_sequential_patterns(sequences, min_support=2):
    patterns = Counter()

    for seq in sequences:
        for i in range(len(seq)):
            for j in range(i + 1, len(seq)):
                patterns[(seq[i], seq[j])] += 1

    return {k: v for k, v in patterns.items() if v >= min_support}
```

```python
patterns = extract_sequential_patterns(customer_sequences.tolist(), min_support=2)
patterns
```
